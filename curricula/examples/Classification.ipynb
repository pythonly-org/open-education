<p data-start="210" data-end="422"><strong data-start="210" data-end="239">What classification does:</strong><br data-start="239" data-end="242">Classification learns patterns from labeled data and predicts the correct class for new samples.<br data-start="340" data-end="343">Here, the goal is to determine the iris species based on flower measurements.</p>
<p data-start="426" data-end="468"><strong data-start="426" data-end="466">How K-Nearest Neighbors (KNN) works:</strong></p>
<ul>
<li>
<ul data-start="471" data-end="702">
<li data-start="471" data-end="503">
<p data-start="473" data-end="503">Stores all training samples.</p>
</li>
<li data-start="506" data-end="581">
<p data-start="508" data-end="581">For each new point, finds the <strong data-start="538" data-end="561">k closest neighbors</strong> in feature space.</p>
</li>
<li data-start="584" data-end="641">
<p data-start="586" data-end="641">Predicts the class most common among those neighbors.</p>
</li>
<li data-start="644" data-end="702">
<p data-start="646" data-end="702">Works well when classes form local clusters in the data.</p>
</li>
</ul>
</li>
</ul>

```python
from pyy.dataset import Dataset
from pyy.plot import ScatterPlot
from pyy.splitter import Splitter
from pyy.noise import Noise
from pyy.classifier import Classifier
from pyy.accuracy_plot import AccuracyPlot

# VISIBLE_CODE_START
#✏️ Classify The Iris Dataset
Dataset(file_name='iris.csv', target=["target"],block_id='~_p$V{M*._ecyw6RG?]a')
ScatterPlot(columns=["sepal length (cm)","sepal width (cm)"], fig_size='6.4 x 4.8', title='(auto)', show_color_bar=True, cmap='tab10',block_id='/CafZ8i;DAIV7c`piXBe')
Splitter(test_size=0.4, random_state=42,block_id='3S/;Uqb|+{qt%[B{;{n0')
Noise(noise_type='uniform', columns=[], min=0.0, max=0.1,block_id='1b*,)VdiK{%R:PcCmPi:')
my_classifier = Classifier(name='K-Nearest Neighbors', n_neighbors=5, weights='uniform', algorithm='auto', metric='minkowski',block_id='Q{437Y_:1f8!=!{AIGPQ')
AccuracyPlot(fig_size='6.4 x 4.8', title='(auto)',block_id='tX9V:R=bI81|?;I3]`6v')
print('Score:' + str(my_classifier.get_score()))
# VISIBLE_CODE_END
```
