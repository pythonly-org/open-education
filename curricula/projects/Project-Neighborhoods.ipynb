<h2>Prediction: Building A Model</h2>
<p data-start="34" data-end="181">To predict the <em data-start="49" data-end="56">match</em>, we&rsquo;ll create a simple model that captures the relationships observed earlier and allows us to make predictions on new data.</p>
<p data-start="183" data-end="203"><strong data-start="183" data-end="203">Model structure:</strong></p>
<ul data-start="205" data-end="390">
<li data-start="205" data-end="285">
<p data-start="207" data-end="285"><strong data-start="207" data-end="235">Encoder &ndash; OneHotEncoder:</strong> used to encode the categorical column <em data-start="274" data-end="282">"name"</em>.</p>
</li>
<li data-start="286" data-end="390">
<p data-start="288" data-end="390"><strong data-start="288" data-end="324">Classifier &ndash; MaxEntropy:</strong> chosen as an example classifier (others could be used as well).</p>
</li>
</ul>
<p data-start="392" data-end="562" data-is-last-node="" data-is-only-node="">These components will be wrapped together inside a <strong data-start="443" data-end="452">Model</strong>, ensuring that the same preprocessing and prediction steps can be consistently applied to any new input data.</p>

```python
from pyy.encoder import Encoder
from pyy.classifier import Classifier
from pyy.model import Model
from pyy.dataset import Dataset

# VISIBLE_CODE_START
my_model=Model(
    Encoder(name='OneHotEncoder', columns=["name"],block_id='(kt[Ql;;?Z*8VDape?m4'),
    Classifier(name='MaxEntropy',block_id='@o:uCMJ~AG-d:C(30`*,')
,block_id='tM`j80YAYsXO%pc#cXb[')

Dataset(file_name='neighborhoods.csv', target=["match"],block_id='5z)yY[LvPSHR.d}=z(Z4')
my_model.fit(data=None)
print(my_model.get_score())
# VISIBLE_CODE_END
```
