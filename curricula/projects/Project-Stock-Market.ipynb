<h2 data-start="0" data-end="22"><span data-sheets-root="1">‚öôÔ∏è</span> Preprocessing</h2>
<p data-start="24" data-end="117">Now that we understand the structure of our dataset, we can define the preprocessing steps:</p>
<ul data-start="119" data-end="400">
<li data-start="119" data-end="185">
<p data-start="121" data-end="185"><strong data-start="121" data-end="139">OrdinalEncoder</strong> &ndash; encodes the <em data-start="154" data-end="162">Ticker</em> and <em data-start="167" data-end="175">Sector</em> columns</p>
</li>
<li data-start="186" data-end="254">
<p data-start="188" data-end="254"><strong data-start="188" data-end="210">DateTimeFeaturizer</strong> &ndash; extracts numerical features from <em data-start="246" data-end="252">Date</em></p>
</li>
<li data-start="255" data-end="333">
<p data-start="257" data-end="333"><strong data-start="257" data-end="267">Scaler</strong> &ndash; applies a <em data-start="280" data-end="296">Min&ndash;Max Scaler</em> to normalize values between (0, 1)</p>
</li>
<li data-start="334" data-end="400">
<p data-start="336" data-end="400"><strong data-start="336" data-end="348">Splitter</strong> &ndash; divides the data into training and testing sets</p>
</li>
</ul>
<p data-start="402" data-end="524" data-is-last-node="" data-is-only-node="">We&rsquo;ll wrap now these steps into a&nbsp;<strong data-start="445" data-end="454">Model</strong> object, allowing us to reuse the entire preprocessing pipeline later.</p>

```python
from pyy.dataset import Dataset
from pyy.encoder import Encoder
from pyy.scaler import Scaler
from pyy.splitter import Splitter
from pyy.model import Model

# VISIBLE_CODE_START
#‚úèÔ∏è After picking the Model, click on its left "output mode" icon to create a variable
Dataset(file_name='stock_market_2025.csv', target=["Close Price"],block_id='+/aPpI.~x~{aznngvX,x')
Model(
    Encoder(name='OrdinalEncoder', columns=["Ticker","Sector"],block_id='k,x.B9wjHmmAVJ-Sj~@D'),
    Encoder(name='DateTimeFeaturizer', columns=["Date"],block_id='rPjrU|uGuq`uyKCKhC1W'),
    Scaler(scaler_type='Min-Max', feature_range=(0,1), columns=[],block_id='hF`,yv|T{:id69ai$GR*'),
    Splitter(test_size=0.4, random_state=42,block_id='i,RMS`Ziw^*gxv90fZI4')
,block_id='#~@Y9$5}N|/[Wl]Zg#9(')
# VISIBLE_CODE_END
```

<h2>Choosing a Classifier / Regressor</h2>
<p data-start="45" data-end="127">Next, we need to select the core of our learning component &mdash; the <strong data-start="110" data-end="124">model type</strong>:</p>
<ul data-start="129" data-end="311">
<li data-start="129" data-end="236">
<p data-start="131" data-end="236"><strong data-start="131" data-end="145">Classifier</strong> &ndash; used for classification tasks where the target column has a limited set of categories.</p>
</li>
<li data-start="237" data-end="311">
<p data-start="239" data-end="311"><strong data-start="239" data-end="252">Regressor</strong> &ndash; used when the target is a continuous numeric variable.</p>
</li>
</ul>
<p data-start="313" data-end="486">Since in our case the <strong data-start="335" data-end="350">Close Price</strong> is a continuous value, we&rsquo;ll use a <strong data-start="386" data-end="399">Regressor</strong>, such as <strong data-start="409" data-end="429">LinearRegression</strong>, which learns by minimizing the sum of squared errors.</p>
<p data-start="488" data-end="550" data-is-last-node="" data-is-only-node="">Let&rsquo;s add <code data-start="498" data-end="516">LinearRegression</code> to our model and print its score.</p>

```python
from pyy.dataset import Dataset
from pyy.encoder import Encoder
from pyy.scaler import Scaler
from pyy.splitter import Splitter
from pyy.regressor import Regressor
from pyy.model import Model
from pyy.accuracy_plot import AccuracyPlot

# VISIBLE_CODE_START
Dataset(file_name='stock_market_2025.csv', target=["Close Price"],block_id='WfVZX!{V:n/K$}6S:dM[')
my_model=Model(
    Encoder(name='OrdinalEncoder', columns=["Ticker","Sector"],block_id='k,x.B9wjHmmAVJ-Sj~@D'),
    Encoder(name='DateTimeFeaturizer', columns=["Date"],block_id='rPjrU|uGuq`uyKCKhC1W'),
    Scaler(scaler_type='Min-Max', feature_range=(0,1), columns=[],block_id='hF`,yv|T{:id69ai$GR*'),
    Splitter(test_size=0.4, random_state=42,block_id='i,RMS`Ziw^*gxv90fZI4'),
    Regressor(name='LinearRegression', fit_intercept=True,block_id='CvilBHb}+faZ/#TukAVa')
,block_id='#~@Y9$5}N|/[Wl]Zg#9(')
print(my_model.get_score())
AccuracyPlot(fig_size='6.4 x 4.8', title='(auto)',block_id='UUU11QOP#B_3t[-#y`,t')

#‚úèÔ∏è Click the "+m" to call the model method
# VISIBLE_CODE_END
```

<p data-start="0" data-end="51">As we can see, the score is very high &mdash; almost 1!</p>
<p data-start="0" data-end="51">This is expected, since several features (like&nbsp;<strong data-start="100" data-end="114">Open Price</strong>) are strongly correlated with <strong data-start="145" data-end="160">Close Price</strong>, making the prediction task relatively easy for the model.</p>

<h2>***Generating and Manipulating Dataset Columns&nbsp;</h2>
<p data-start="54" data-end="186">The <strong data-start="58" data-end="65">Map</strong> block (found in the <em data-start="86" data-end="94">Filter</em> category) lets us create or modify dataset columns by applying an expression to each row.</p>
<p data-start="188" data-end="220">It takes three main arguments:</p>
<ul data-start="221" data-end="652">
<li data-start="221" data-end="496">
<p data-start="223" data-end="385"><strong data-start="223" data-end="231">expr</strong> &ndash; a free-text expression defining how to compute the new value.<br data-start="295" data-end="298">You can access columns using <strong>self.&lt;column-name&gt;</strong> or <strong>self[&lt;column-name&gt;]</strong> and apply standard pandas operations such as:</p>
<ul data-start="388" data-end="496">
<li data-start="388" data-end="422">
<p data-start="390" data-end="422">Basic math: <code data-start="402" data-end="405">+</code>, <code data-start="407" data-end="410">-</code>, <code data-start="412" data-end="415">*</code>, <code data-start="417" data-end="420">/</code></p>
</li>
<li data-start="425" data-end="496">
<p data-start="427" data-end="496">Aggregate functions: <code data-start="448" data-end="455">sum()</code>, <code data-start="457" data-end="465">mean()</code>, <code data-start="467" data-end="476">count()</code>, <code data-start="478" data-end="485">max()</code>, <code data-start="487" data-end="494">min()</code></p>
</li>
<li data-start="497" data-end="575">Comparison: &gt;, &gt;=, &lt;, &lt;=, ==, !=</li>
</ul>
</li>
<li data-start="497" data-end="575">
<p data-start="499" data-end="575"><strong data-start="499" data-end="513">set_target</strong> &ndash; whether the new column should become the target variable.</p>
</li>
<li data-start="576" data-end="652">
<p data-start="578" data-end="652"><strong data-start="578" data-end="585">out</strong> &ndash; the name of the output column (can overwrite an existing one).</p>
</li>
</ul>
<p data-start="654" data-end="769" data-is-last-node="" data-is-only-node="">For example, the following expression creates a new column <code data-start="713" data-end="725">birth_year</code> by subtracting <code data-start="741" data-end="746">age</code> from the current year:</p>

<h2 data-start="0" data-end="67">Stock Market Dataset: Generating the <em data-start="44" data-end="58">Up Direction</em> Column</h2>
<p data-start="69" data-end="176">We can now easily add a new boolean column indicating whether a stock&rsquo;s price went <strong data-start="152" data-end="158">up</strong> during the day.</p>
<p data-start="69" data-end="176">We will apply<strong data-start="293" data-end="314"> </strong>a<strong data-start="293" data-end="314"> Map </strong>block to create the <code data-start="329" data-end="343">Up Direction</code> column by subtracting the <strong data-start="370" data-end="384" data-is-only-node="">Open Price</strong> from the <strong data-start="394" data-end="409">Close Price</strong>:&nbsp;</p>
<ul>
<li data-start="69" data-end="176">If the result is positive &rarr; the stock went <strong data-start="461" data-end="467">up</strong></li>
</ul>
<p><strong>Note</strong>: since the column names have space included, we'll use this time the array notation, e.g. self['Open Price']</p>

```python
from pyy.dataset import Dataset
from pyy.map_data import Map
from pyy.table_view import TableView

# VISIBLE_CODE_START
Dataset(file_name='stock_market_2025.csv', target=[],block_id='/lO4TUatTG5MGk6S16bu')
Map(expr="""self["Close Price"]-self["Open Price"] > 0""", out='Up Direction', set_target=True,block_id='[eEFr~va5BT=]l^_(p:x')
TableView(max_rows=10, max_cols=10, title='(auto)', has_header=True, has_border=True, transpose=False,block_id='5]*f_n7Q9@jgQ_:%9;4P')
# VISIBLE_CODE_END
```

<p>Now let's combine that with our previous model, with now replacing the regression model with a classifier: <strong>MaxEntropy</strong></p>

```python
from pyy.dataset import Dataset
from pyy.map_data import Map
from pyy.drop import Drop
from pyy.encoder import Encoder
from pyy.scaler import Scaler
from pyy.splitter import Splitter
from pyy.classifier import Classifier
from pyy.model import Model
from pyy.accuracy_plot import AccuracyPlot

# VISIBLE_CODE_START
Dataset(file_name='stock_market_2025.csv', target=[],block_id='Jw9m6e6IH{y(3[drvcO-')
my_model=Model(
    Map(expr="""cols["Close Price"]-cols["Open Price"] > 0""", out='Up Direction', set_target=True,block_id='[eEFr~va5BT=]l^_(p:x'),
    Drop(columns=["Close Price"],block_id='/trbt{uDwF^?.Gnk?B|n'),
    Encoder(name='OrdinalEncoder', columns=["Ticker","Sector"],block_id='k,x.B9wjHmmAVJ-Sj~@D'),
    Encoder(name='DateTimeFeaturizer', columns=["Date"],block_id='rPjrU|uGuq`uyKCKhC1W'),
    Scaler(scaler_type='Min-Max', feature_range=(0,1), columns=[],block_id='hF`,yv|T{:id69ai$GR*'),
    Splitter(test_size=0.2, random_state=42,block_id='i,RMS`Ziw^*gxv90fZI4'),
    Classifier(name='MaxEntropy',block_id='CvilBHb}+faZ/#TukAVa')
,block_id='#~@Y9$5}N|/[Wl]Zg#9(')
print(my_model.get_score())
AccuracyPlot(fig_size='6.4 x 4.8', title='(auto)',block_id='ImfKq0GhB1x7HeOYux]/')
# VISIBLE_CODE_END
```

<h2 data-start="108" data-end="136">üìä&nbsp;üß© Results Explanation</h2>
<p data-start="138" data-end="423">Why is the score now just above 50%, while the regression model scored much higher? is that a bug??</p>
<p data-start="138" data-end="423">In the regression task, the model predicted the&nbsp;<strong data-start="272" data-end="294">actual close price</strong>, which changes smoothly and is strongly correlated with the open and previous close &mdash; so even simple models achieve low error.</p>
<p data-start="425" data-end="615">In the classification task, we ask the model to predict the <strong data-start="485" data-end="498">direction</strong> (up or down).<br data-start="512" data-end="515">This is far harder, since small, random fluctuations or unseen market factors determine that sign.</p>
<p data-start="617" data-end="754">Thus, while regression captures general price levels (a mostly predictable trend),<br data-start="699" data-end="702">the direction prediction approaches random chance.</p>
<p data-start="756" data-end="864" data-is-last-node="" data-is-only-node="">A 0.55&ndash;0.60 accuracy is therefore realistic &mdash; it means the model finds weak but genuine patterns, not a bug.</p>
<p></p>
