<h2>Expressions</h2>
<p>Expressions combine values and operators to produce new results.</p>
<p>You can think of them as questions you ask Python to compute.</p>

```python

# VISIBLE_CODE_START
your_age = age + 5
average_temp = (temperature + 22) / 2
weekend_message = 'hello'
if your_age >= 18:
  print("You're now responsible for yourself!")
# VISIBLE_CODE_END
```

<p>Here we used arithmetic (`+`, `/`) and conditional (`if ... `) expressions.</p>
<p>Each produces a value that can be assigned to a variable or printed.</p>

<h2>Conditional Logic</h2>
<p>Programs can choose what to do depending on conditions using `if`, `elif`, and `else` statements.</p>

```python

# VISIBLE_CODE_START
if temperature > 25:
  print("It's a hot day!")
else:
  print('Nice weather outside.')
# VISIBLE_CODE_END
```

<p>Only one branch runs: the first whose condition is `True`.</p>
<p>Here we compare the numeric variable `temperature` and print a different message depending on the result.</p>

<h3>Combining Conditions</h3>
<p>You can combine multiple comparisons with logical operators `and`, `or`, and `not` to express more complex rules.</p>

```python

# VISIBLE_CODE_START
#‚úèÔ∏è Use the Logic category to drag these blocks in
if is_weekend and temperature > 15:
  print('Perfect day for a walk!')
else:
  print('Maybe stay and read a book.')
# VISIBLE_CODE_END
```

<p>`and` requires both conditions to be true, `or` requires at least one, and `not` reverses a boolean value.</p>
<p>Logical combinations allow you to express real-world decision rules.</p>

<h2>Mini Project: Smart Alarm</h2>
<p>Now that you can handle logic, let's combine it all in a short program.</p>
<p>We'll decide whether to wake up or sleep in based on the day and time.</p>

```python

# VISIBLE_CODE_START
#‚úèÔ∏è Click '<>' to view the generated Python code
day = 'Monday'
hour = 7
if day in ["Saturday", "Sunday"] or hour < 7:
  print('Shhhh...')
else:
  print('Wake up!')
# VISIBLE_CODE_END
```

<p>We check both conditions at once: the day of week and the time.</p>
<p>Because if either is true we keep sleeping, we use `or`.</p>
<p>Lists make it easy to group weekend names together for the test: <code>day in ["Saturday", "Sunday"]</code></p>
